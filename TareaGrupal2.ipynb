```python
import pandas as pd 
```

```python
dataForest="https://www.cia.gov/the-world-factbook/field/revenue-from-forest-resources/country-comparison" 
```

```python
forest=pd.read_html(dataForest) #escarpear: se va ezportar todas las listas de la página 
forest #esto refiere a un link a la página, es decir muchas tablas 
```

```python
type(forest) #me señala que tipo de data es forest 
```

```python
forest[0] #Utilizando el [0] se escoge la primera lista
```

```python
#1. Replace el % por pct
forestOK=forest[0].rename(columns={"% of GDP":"pct_of_GDP"})
```

```python
forestOK #se muestra el cambio de nombre de una columna 
```

```python
#2. Mantener las columnas de Country, pct of GDP, and Date of Information
byeColumns=['Rank']
```

```python
forestOK.drop(columns=byeColumns,inplace=True)
```

```python
forestOK #se actualizó la tabla sin la columna de Rank 
```

```python
#3. Cambiar el nombre de la columna Date of Information por forest_date
forestOK=forestOK.rename(columns={"Date of Information":"forest_date"})
```

```python
forestOK
```

```python
#6. Mantener solo el año en la columna forest date 
forestOK.forest_date=forestOK.forest_date.str.split(' ',expand=True)[0]
```
